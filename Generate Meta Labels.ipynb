```python
import pandas as pd
```

Read output from primary model!

```python
predictions_path = './data/output.pkl'
with open(predictions_path,'rb') as file:
    predicted_labels = pd.read_pickle(file)
```

```python
num_stocks = 93

# Initialize a list to hold the extracted data for each stock
extracted_data = [[] for _ in range(num_stocks)]

# Extract data for each stock
for stock_index in range(num_stocks):
    extracted_data[stock_index] = [array[stock_index] for array in predicted_labels]

# Now extracted_data[i] contains all the data for stock i
print(extracted_data)
```

Read acutal labels from the processed data

```python
stock_name = 'وسکاب'

processed_data_path = './data/processed_data/stock_with_indcor/' + stock_name + '.csv'
data = pd.read_csv(processed_data_path)
```

```python
df = data[-65:]
```

```python
df['signal'] = extracted_data[83]
```

```python
df
```

```python

# Assuming you have a DataFrame 'df' with the following columns:
# - 'primary_prediction': the predictions made by the primary model
# - 'actual_outcome': the actual outcomes (1 for success, 0 for failure)

# Create a new column for meta-labels
df['meta_label'] = 0

# Generate meta-labels based on the primary model's predictions and actual outcomes
df.loc[(df['signal'] == 1) & (df['label'] == 1), 'meta_label'] = 1
df.loc[(df['signal'] == 1) & (df['label'] == 0), 'meta_label'] = 0

# Print the updated DataFrame
print(df[['signal', 'label','meta_label']].head(-10))
```

```python
data = df[['Volume_rate','high_rate','low_rate','open_rate','earning_rate','individual_buy_count','corporate_buy_count','individual_sell_count','corporate_sell_count','individual_buy_vol','corporate_buy_vol','individual_sell_vol','corporate_sell_vol','individual_buy_value','corporate_buy_value','individual_sell_value','corporate_sell_value','individual_buy_mean_price','individual_sell_mean_price','corporate_buy_mean_price','corporate_sell_mean_price', 'meta_label']]
```
