```python
import pandas as pd
import numpy as np
import random
```

Read model predictions for 65 days for all stocks

```python
predictions_path = './data/output.pkl'
with open(predictions_path,'rb') as file:
    predicted_labels = pd.read_pickle(file)
predicted_labels.shape
```

Choose 5 random stock to evaluate the predictions

```python
random.seed(1)
random_numbers = random.sample(range(94), 10)
random_numbers
```

```python
# Number of stocks
num_stocks = 93

# Initialize a list to hold the extracted data for each stock
extracted_data = [[] for _ in range(num_stocks)]

# Extract data for each stock
for stock_index in range(num_stocks):
    extracted_data[stock_index] = [array[stock_index] for array in predicted_labels]

# Now extracted_data[i] contains all the data for stock i
print(extracted_data)
```

Read stock raw data and extract 65 last day

```python
stock_name = 'وسکاب'
data = pd.read_csv('./data/raw_data/stocks/'+stock_name+'.csv')
subset_data = data[['Date','Open','High','Low','Close','Volume']][-65:]
```

Merge the predictions and the stock HLOCV
